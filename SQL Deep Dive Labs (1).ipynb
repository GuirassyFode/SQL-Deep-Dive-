**NOT IN**

select * from CUSTOMERS WHERE  CustomerID  NOT IN (SELECT CustomerID FROM Orders); 

**NOT EXISTS**

select * from CUSTOMERS C WHERE NOT EXISTS (SELECT * FROM Orders o where C.CustomerID = o.CustomerID)

**EXISTS**

select * from CUSTOMERS C WHERE EXISTS (SELECT * FROM Orders o where C.CustomerID = o.CustomerID)

**JOIN**

select *  from CUSTOMERS AS C LEFT JOIN ORDERS O ON C.CustomerID=O.CustomerID WHERE O.CustomerID IS NULL;
